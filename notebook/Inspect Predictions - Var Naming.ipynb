```python
n=7; i=600
preds_all = []; preds = []
for idx, row in results_df.iterrows():
    top_n = list(np.argsort(-row)[:n])
    preds.append(top_n[:n])
    #pred = np.asarray(top_n[:n])
    if (idx % 64 == 63) and (idx > 0):
        preds = np.asarray(preds)
        last_idx=64
        #print(idx // 64, ' '.join([vocab_label_df.loc[p][0] for p in preds[:64,0]]))
        #print("Label = ", ' '.join(list(label_df.loc[idx//64][:4])))
        preds_all.append(preds)
        preds = []
        #print()
    if idx > i*64:
        break
```

```python
labels = []; labels_str = []
for idx, row in label_df.iterrows():
    l = vocab_label_df.index[vocab_label_df[0]==row[0]]
    if len(l) > 0:
        labels.append(l[0])
    else:
        labels.append(-1)
    labels_str.append(list(row[:-1]))
```

```python
snippet = pd.read_csv(path+'sparse_varname_split_magret_tk_val.txt', header=None)
snippet.shape
```

```python
n=1
pred_str = []; score = 0; score_no_pad=0; rank =[]; skipped = 0
for idx in range(600):
    print(idx)
    print(snippet.loc[idx][0])
    print("Label =", labels_str[idx])
    try:
        msk_idx = snippet.loc[idx][0].split(" ").index('[MASK]')
    except:
        msk_idx = -1
        skipped +=1
    
    if msk_idx == -1:
        continue
    preds_ = []
    print("Pred  =")
    r = preds_all[idx]
    for i in range(n):
        p = [vocab_label_df.loc[r[msk_idx+j][i]][0] for j in range(4)]
        for k,p_ in enumerate(p):
            print(p_, labels_str[idx][k])
            if p_==labels_str[idx][k]:
                score +=1
                if p_ != '[PAD]':
                    score_no_pad +=1
        print("     {}. {}".format(i,p))
        preds_.append(p)
    pred_str.append(preds_)
    print()
```

```python
n=1; nb_snips = 600
pred_str = []; score = [0]*4; score_full_name=0; score_no_pad = 0; rank =[0]*4; skipped = 0
for idx in range(nb_snips):
    print(idx)
    print(snippet.loc[idx][0])
    print("Label =", labels_str[idx])
    try:
        msk_idx = snippet.loc[idx][0].split(" ").index('[MASK]')
    except:
        msk_idx = -1
        skipped +=1

    if msk_idx == -1:
        continue
    preds_ = []
    print("Pred  =")
    r = preds_all[idx]
    
    label_len = 0; full_token_score = 0
    for j in range(4):
        l = labels_str[idx][j]
        if l != "[PAD]":
            label_len += 1
        for i in range(n):
            p = vocab_label_df.loc[r[msk_idx+j][i]][0]
            print(p,l)
            if p == l:
                score[j] += 1
                rank[j] += (i+1)
                if l != "[PAD]":
                    full_token_score += 1
                break
            elif i==n-1:
                rank[j] += n
    if full_token_score == label_len:
        score_full_name += 1
    score_no_pad += full_token_score / label_len
    
    print()
```

```python
[(nb_snips-skipped) / _ for _ in rank]
```
